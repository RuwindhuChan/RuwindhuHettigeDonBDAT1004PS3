```python
## Step 4. Discover what is the mean age per occupation
mean_age = users.groupby(users['occupation'])['age'].aggregate('mean')
mean_age
```

```python
## Step 5. Discover the Male ratio per occupation and sort it from the most to the least
group_by_occupations = users[['occupation','gender']].groupby('occupation').count()
group_by_occupations
```

```python
males_occupations = users.loc[users['gender'] == 'M'][['gender','occupation']]

males_group_by_occupations = males_occupations.groupby('occupation').count()

males_group_by_occupations

```

```python
males_ratio =  males_group_by_occupations / group_by_occupations

males_ratio.sort_values(ascending= False, by = 'gender')
```

```python
## Step 6. For each occupation, calculate the minimum and maximum ages

min_age_occupation = users.groupby(users['occupation'])['age'].aggregate('min')
min_age_occupation
```

```python
max_age_occupation = users.groupby(users['occupation'])['age'].aggregate('max')
max_age_occupation
```

```python
# Step 7. For each combination of occupation and sex, calculate the mean age
mean_age_occ_sex = users.groupby(['occupation','gender'])['age'].aggregate('mean')
mean_age_occ_sex
```

```python
#Step 8. For each occupation present the percentage of women and men

total_counts = users.groupby(['occupation'])['zip_code'].aggregate('count')
occupation_counts = users.groupby(['occupation','gender'])['zip_code'].aggregate('count')
occupation_percent = round((occupation_counts / total_counts) *100,2)
occupation_percent
```

```python
## Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.
data['date_time'] = pd.to_datetime(data[["Yr","Mo","Dy"]].astype(str).agg('-'.join, axis=1))

data.drop(data.iloc[:, 0:3], inplace=True, axis=1)
```

```python
## Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
pd.DatetimeIndex(data['date_time']).year.unique()
```

```python
def fix_century(date_time):
    if date_time.year > 1979:
        year=date_time.year-100 
    else:
        year = date_time.year
    return datetime.date(year,date_time.month,date_time.day)
data['date_time']=data['date_time'].apply(fix_century)
data
```

```python
## Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
type(data['date_time'][0])
```

### Question 7 
Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.

```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as mp

us_marriages = pd.read_csv('us-marriages-divorces-1867-2014.csv')

us_marriages = us_marriages.loc[us_marriages['Year'].isin([1900,1950,2000])]
```

```python
marriage_per_capita = us_marriages['Marriages_per_1000']
marriage_per_capita = marriage_per_capita.values
divorce_per_capita = us_marriages['Divorces_per_1000']
divorce_per_capita = divorce_per_capita.values
year = us_marriages['Year']

df = pd.DataFrame({'Marriages_Per_Capita':marriage_per_capita, 'Divorces_Per_Capita':divorce_per_capita}, index=year)

df.plot.bar()

mp.xlabel("Year")

mp.ylabel("Marriages/Divorces Per Capita")

mp.legend(["Marriages Per Capita", "Divorces Per Capita"])

mp.show()
```
